Ingestão de Dados

```python
import pandas as pd
dados = pd.read_excel('/content/planilha_transporte_2024.xlsx')
print(dados.head())
```

Tratamento de Dados

```python
dados['Data da Viagem'] = pd.to_datetime(dados['Data da Viagem'], dayfirst=True)
dados['Dia da semana'] = dados['Dia da semana'].str.title().str.strip()
dados['Destino'] = dados['Destino'].str.title().str.strip()
dados['Tipo de Carga'] = dados['Tipo de Carga'].str.title().str.strip()
dados['Método de Pagamento'] = dados['Método de Pagamento'].str.title().str.strip()

print(dados.head())

dados['Lucro Bruto (R$)'] = dados['Valor do Frete (R$)'] - dados['Custo Total da Operação (R$)']
dados['Margem (%)'] = ((dados['Lucro Bruto (R$)'] / dados['Valor do Frete (R$)']) * 100).round(2)

print(dados.head())

dados['Mês'] = dados['Data da Viagem'].dt.month_name()
dados['Data Formatada'] = dados['Data da Viagem'].dt.strftime('%d/%m/%Y')

traduzir_mes = {
    'January': 'Janeiro', 'February': 'Fevereiro', 'March': 'Março',
    'April': 'Abril', 'May': 'Maio', 'June': 'Junho',
    'July': 'Julho', 'August': 'Agosto', 'September': 'Setembro',
    'October': 'Outubro', 'November': 'Novembro', 'December': 'Dezembro'
}

dados['Mês'] = dados['Mês'].map(traduzir_mes)

print(dados[['Data da Viagem', 'Data Formatada', 'Mês']].head())

print("Amostra dos dados tratados:")
print(dados[['Data Formatada', 'Mês', 'Destino', 'Tipo de Carga', 'Lucro Bruto (R$)', 'Margem (%)']].head())

analise_mes = dados.groupby('Mês')[['Lucro Bruto (R$)', 'Margem (%)']].mean().reset_index()
print("Lucro e Margem Média por Mês:")
print(analise_mes.sort_values(by='Mês'))

analise_tipo = dados.groupby('Tipo de Carga')[['Lucro Bruto (R$)', 'Margem (%)']].mean().reset_index()
print("Lucro e Margem Média por Tipo de Carga:")
print(analise_tipo.sort_values(by='Lucro Bruto (R$)', ascending=False))

analise_destino = dados.groupby('Destino')[['Lucro Bruto (R$)']].sum().sort_values(by='Lucro Bruto (R$)', ascending=False).head(10)
print("Top 10 Destinos Mais Lucrativos:")
print(analise_destino)

pagamentos = dados['Método de Pagamento'].value_counts()
print("Métodos de Pagamento Mais Utilizados:")
print(pagamentos)

dados_refinados = dados[[
    'Data Formatada', 'Mês', 'Destino', 'Tipo de Carga',
    'Peso da Carga (kg)', 'Valor do Frete (R$)', 'Custo Total da Operação (R$)',
    'Lucro Bruto (R$)', 'Margem (%)', 'Método de Pagamento'
]]

dados_refinados.to_excel("dados_tratados_transporte.xlsx", index=False)

print("Arquivo salvo como: dados_tratados_transporte.xlsx")
print("Prévia dos dados exportados:")
print(dados_refinados.head())
```

Dado Simulado

```python
dados['Custo Elétrico (R$)'] = (dados['Custo Total da Operação (R$)'] * 0.70).round(2)

dados['Lucro Elétrico (R$)'] = (dados['Valor do Frete (R$)'] - dados['Custo Elétrico (R$)']).round(2)

dados['Margem Elétrica (%)'] = ((dados['Lucro Elétrico (R$)'] / dados['Valor do Frete (R$)']) * 100).round(2)

print("\nSimulação de operação com caminhão elétrico:")
print(dados[['Valor do Frete (R$)', 'Custo Total da Operação (R$)', 'Custo Elétrico (R$)', 'Lucro Bruto (R$)', 'Lucro Elétrico (R$)', 'Margem (%)', 'Margem Elétrica (%)']].head())

dados.to_excel("dados_transporte_com_caminhao_eletrico.xlsx", index=False)
print("\nArquivo salvo como: dados_transporte_com_caminhao_eletrico.xlsx")
```
